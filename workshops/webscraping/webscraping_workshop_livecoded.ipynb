# Scraping US Congress

```python
from bs4 import BeautifulSoup
import requests
```

```python
url =  "https://www.house.gov/representatives"
```

```python
text = requests.get(url).text
```

```python
soup = BeautifulSoup(text, parser="html5lib")
```

```python
link_list = []
for a in soup('a'):
    if a.has_attr('href'):
        link_list.append(a)
```

```python
link_list = [a['href'] for a in soup('a') if a.has_attr('href')]
```

```python
len(link_list)
```

```python
link_list
```

```python
re_pattern = r"^https?:\/\/.*\.house\.gov\/?$"
```

```python
import re
```

```python
good_urls = [url for url in link_list if re.match(re_pattern, url)]
```

```python
good_urls
```

```python
better_urls = list(set(good_urls))
```

```python
len(better_urls)
```

```python
text = requests.get("https://pence.house.gov").text
soup = BeautifulSoup(text, "html5lib")

# gather from this soup all the href's that have "press releases"

links = {a['href'] for a in soup('a') if "press releases" in a.text.lower()}
```

```python
links
```

```python
from tqdm.notebook import tqdm

press_releases = {}
# let's find out all the webpages of press releases of all the members of congress
for house_url in tqdm(better_urls):
    text = requests.get(house_url).text
    soup = BeautifulSoup(text, "html5lib")
    # get the press release link from that member
    pr_link = {a['href'] for a in soup('a') if "press releases" in a.text.lower()}
    # print(f"{house_url} : {pr_link}")
    press_releases[house_url] = pr_link
    
```

```python
text = requests.get("https://neal.house.gov/press-releases").text
soup = BeautifulSoup(text, "html5lib")
paragraphs = [p.text for p in soup('p')]
```

```python
def paragraph_mentions(text, keyword):
    """
    returns True if keyword in text
    """
    soup = BeautifulSoup(text, "html5lib")
    paragraphs = [p.get_text() for p in soup('p')]
    
    return any(keyword in paragraph for paragraph in paragraphs)
```

```python
paragraph_mentions(text, "Springfield")
```

```python
# loop through press_releases dictionary
# rebuild the url: congressmember-URL + press-release-URL(s)
# "soup" that URLs
# inspect the snippets or abstracts of the press releases for our keyword
# if keyword found, add url to list

def what_is_mentioned_by_which_congressmember(press_releases, keyword):
    url_list = []
    for house_url, pr_links in tqdm(press_releases.items()):
        for pr_link in pr_links:
            url = f"{house_url}/{pr_link}"

            text = requests.get(url).text

            if paragraph_mentions(text, keyword):
                print(f"{house_url}{pr_link}")
                return url_list.append(f"{house_url}{pr_link}")
                break   # leave the for loop, go to next house_url
```

```python
cm_who_talk_about_socialism = what_is_mentioned_by_which_congressmember(press_releases, "socialism")
```

```python
cm_who_talk_about_blm = what_is_mentioned_by_which_congressmember(press_releases, "Black lives matter")
```

```python
cm_who_talk_about_election_fraud = what_is_mentioned_by_which_congressmember(press_releases, "election fraud")
```
